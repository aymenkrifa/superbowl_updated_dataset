```python
import pandas as pd
```

```python
super_bowls = pd.read_csv('old_super_bowls.csv')
super_bowls.head()
```

```python
df = pd.read_csv("old_super_bowl2.csv")

df.columns=["date", "super_bowl", "winner", "winner_points", "loser", 
            "loser_points", "mvp_player", "stadium", "city", "state"]

df.head()
```

```python
df.insert(6, "attendance", [24835, 62417, 70081] + super_bowls["attendance"].tolist())
df.insert(6, "difference_points", abs(df["winner_points"] - df["loser_points"]))
df.insert(6, 'combined_points', df["winner_points"] + df["loser_points"])
df.insert(6, "loser_coach", ["Andy Reid", "Kyle Shanahan", "Sean McVay"] + super_bowls["coach_loser"].tolist())
df.insert(6, "winner_coach", ["Bruce Arians", "Andy Reid", "Bill Belichick"] + super_bowls["coach_winner"].tolist())

df.insert(6, "winner_qb", ["Tom Brady", "Patrick Mahomes", "Tom Brady"] + super_bowls["qb_winner_1"].tolist())
df.insert(6, "loser_qb", ["Patrick Mahomes", "Jimmy Garoppolo", "Jared Goff"] + super_bowls["qb_loser_1"].tolist())

df.head()
```
